```python
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
fig,ax = plt.subplots()
y = [0.06,0.18,0.21,0.53,0.86,1.70]
x= [50,60,80,100,150,200]

sns.pointplot(x=x,y=y)
plt.xlabel("Runtime S")
plt.ylabel("Lattice Size")
plt.title("Cython_4Cores vs Lattice Size")
```

```python
fig,ax = plt.subplots()
y = [0.18,0.08,0.065,0.113]
x= [1,2,4,8]

sns.pointplot(x=x,y=y)
plt.xlabel("Cores")
plt.ylabel("Speed (S)")
plt.title("Cython_Cores vs Speed")
```
